# Установка интерпретатора и пакетного менеджера (1 балл)

В современном мире языки программирования сами по себе не так интересны, а интересны инструменты, на них написанные. Эти инструменты называются пакетами или библиотеками, и нам нужно научиться их себе ставить. Для этого есть нечто, что называется пакетными менеджерами, которыми нужно научиться пользоваться. Пакетных менеджеров несколько, но наиболее распостранены два: `pip`, который идет как приложение к питону, и `conda` в составе анаконды. В данный момент разницы между ними не так много, так что выбирайте любой.

## Установка пакетов через `pip`

Установка питона без анаконды на самом деле не такое сложное дело. Тут нужно пройти всего несколько шагов:
1. Зайти на сайт https://www.python.org/.
2. Навести мышью на `Downloads` в меню, выбрать пункт `Windows`. Если у вас другая система -- её соответственно (если нужно -- сделаем инструкции и там тоже).
3. Дальше там будет две колонки, в левой из них будет заголовок `Stable releases`. Среди всех версий нас интересует 3.8.10. Почему она? Она работает под Windows 7. Находим ее чуть ниже в списке, переходим по ссылке [Windows installer (64-bit)](https://www.python.org/ftp/python/3.8.10/python-3.8.10-amd64.exe), загружаем инсталлятор.
4. На первом экране инсталлятора не забываем поставить галочку `Add Python 3.8 to PATH` в самом низу окна, после чего выбираем пункт  `Install Now`. 
5. Теперь запустим нечто, что называется терминалом. Это такой же инструмент общения с операционной системой, как и графический *проводник* Windows (который предоставляет рабочий стол, меню пуск и окошки с папками). Возможно он покажется непонятным на первый взгляд, но это временно. Для запуска терминала зажмите одновременно `Win+R` (третью кнопку слева в нижнем ряду с флагом windows, а еще клавишу R). Введите там `powershell`, после чего нажмите клавишу `Enter`. Если у вас открылось синее окно с текстом, то вы все сделали верно.
6. Теперь мы можем устанавливать необходимые для работы пакеты. На первое время нам понадобятся только два: `notebook` и `pytest`. Для их установки выполните в терминале следующую команду `pip install notebook pytest --user`. Здесь `pip`  -- пакетный менеджер питона, который нужен для установки/удаления/контроля сторонних питоновских библиотек. `install` -- указание установить все пакеты, которые будут указаны после. Ну а после уже идут названия необходимых нам пакетов, `notebook` и `pytest`. Никакой магии тут нет, все максимально прозрачно.
